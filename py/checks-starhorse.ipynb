```python
# load in the correct modules
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import tqdm
import cmasher as cm
import warnings
warnings.filterwarnings('ignore')
from matplotlib.colors import LogNorm
from matplotlib import rc
# rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('font',**{'family':'serif','serif':['Times']})
rc('text', usetex=True)
```

```python
# LOAD THE DATA
path = '/Users/dhortadarrington/Documents/Projects/aging-the-sausage/dat/intermediate-age-LAMOST-StarHorse.fits'
tb = fits.open(path)
data = tb[1].data 
```

```python
plt.figure(figsize=(16,14), facecolor='white')

plt.subplot(3,3,1)
plt.title('Mass', fontsize=20)
plt.plot([0.75,1.05], [0.75,1.05], lw=2, ls='dashed')
plt.scatter(data['Mass'], data['Mass50'], s=100)
# plt.xlabel('Xiang and Rix', fontsize=20)
plt.ylabel('Starhorse', fontsize=20)
plt.xlim(0.75,1.05)
plt.ylim(0.75,1.05)

plt.subplot(3,3,2)
plt.title('Age', fontsize=20)
plt.plot([6,14], [6,14], lw=2, ls='dashed')
plt.scatter(data['Age'], data['Age50'], s=100)
# plt.xlabel('Xiang and Rix', fontsize=20)
plt.ylabel('Starhorse', fontsize=20)
plt.xlim(6,14)
plt.ylim(6,14)

plt.subplot(3,3,3)
plt.title('Temperature', fontsize=20)
plt.plot([5300,6300], [5300,6300], lw=2, ls='dashed')
plt.scatter(data['TEFF'], data['Teff50'], s=100)
plt.xlabel('Xiang and Rix', fontsize=20)
plt.ylabel('Starhorse', fontsize=20)
plt.xlim(5300, 6300)
plt.ylim(5300, 6300)

plt.subplot(3,3,4)
plt.title('LOGG', fontsize=20)
plt.plot([3.5,4.3], [3.5,4.3], lw=2, ls='dashed')
plt.scatter(data['LOGG'], data['logg50'], s=100)
plt.xlabel('Xiang and Rix', fontsize=20)
plt.ylabel('Starhorse', fontsize=20)
plt.xlim(3.5, 4.3)
plt.ylim(3.5, 4.3)

plt.subplot(3,3,5)
plt.title('Metallicity', fontsize=20)
plt.plot([-1.8,-0.6], [-1.8,-0.6], lw=2, ls='dashed')
plt.scatter(data['FEH'], data['Met50'], s=100)
plt.xlabel('Xiang and Rix', fontsize=20)
plt.ylabel('Starhorse', fontsize=20)
plt.xlim(-1.8,-0.6)
plt.ylim(-1.8,-0.6)

# plt.subplot(3,3,6)
# plt.title('LOGG', fontsize=20)
# plt.plot([3.5,4.3], [3.5,4.3], lw=2, ls='dashed')
# plt.scatter(data['LOGG'], data['logg50'], s=100)
# plt.xlabel('Xiang and Rix', fontsize=20)
# plt.ylabel('Starhorse', fontsize=20)
# plt.xlim(3.5, 4.3)
# plt.ylim(3.5, 4.3)

```
